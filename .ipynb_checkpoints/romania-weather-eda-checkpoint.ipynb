```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore', category=FutureWarning)
```

```python
df = pd.read_csv("/kaggle/input/romania-weather-visual-crossing-weather/weather_2011-2021_Romania.csv" , encoding='latin-1')
df.head()
```

```python
df.isna().sum()
```

```python
df.drop(['Heat Index', 'Info', 'Wind Chill'], axis=1, inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
df['Date time'] = pd.to_datetime(df['Date time'], format='%m/%d/%Y')
```

```python
df['Temperature Range'] = df['Maximum Temperature'] - df['Minimum Temperature']
df['Day of Week'] = df['Date time'].dt.day_name()
df['Month'] = df['Date time'].dt.month
df['Year'] = df['Date time'].dt.year
df['Season'] = df['Date time'].dt.month%12 // 3 + 1
df['Season'] = df['Season'].map({1: 'Winter', 2: 'Spring', 3: 'Summer', 4: 'Fall'})
```

```python
df['Normalized Wind Speed'] = (df['Wind Speed'] - df['Wind Speed'].mean()) / df['Wind Speed'].std()
df['Wind Direction Category'] = pd.cut(df['Wind Direction'], bins=[0, 90, 180, 270, 360], 
                                       labels=['North', 'East', 'South', 'West'], right=False)
```

```python
common_suffix = ', România'
df['Resolved Address'] = df['Resolved Address'].str.replace(common_suffix, '', regex=False)
df['Name'] = df['Name'].str.replace(common_suffix, '', regex=False)
```

```python
df.head()
```

```python
df.to_csv('cleaned_romania_weather_dataset.csv', index=False)
```

```python
monthly_temps = df.groupby('Month')['Temperature'].mean()

# Lineplot for the averae monthly temperature
plt.figure(figsize=(10, 5))
sns.lineplot(data=monthly_temps)
plt.title('Average Monthly Temperature')
plt.xlabel('Month')
plt.ylabel('Average Temperature (°C)')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(True)
plt.show()
```

```python
weather_counts = df['Weather Type'].value_counts()

# Barplot for top 10 most frequent weather conditions
plt.figure(figsize=(12, 8))
barplot =  sns.barplot(x=weather_counts.head(10).values, y=weather_counts.head(10).index)
plt.title('Top 10 Weather Condition Combinations')
plt.xlabel('Frequency')
plt.ylabel('Weather Condition Combinations')

for p in barplot.patches:
    width = p.get_width()    
    plt.text(width + 3,      
             p.get_y() + p.get_height() / 2,  
             '{:1.0f}'.format(width),         
             ha = 'left',   
             va = 'center') 

plt.show()
```

```python
# Densityplot of Temperature vs Relative Humidity
plt.figure(figsize=(10, 6))
sns.kdeplot(x=df['Temperature'], y=df['Relative Humidity'], cmap='Blues', fill=True)
plt.title('2D Density Plot of Temperature vs. Relative Humidity')
plt.xlabel('Temperature (°C)')
plt.ylabel('Relative Humidity (%)')
plt.show()
```

```python
yearly_visibility = df.groupby('Year')['Visibility'].mean().reset_index()
# Barplot for the average yearly visibility
plt.figure(figsize=(10, 5))
sns.barplot(x='Year', y='Visibility', data=yearly_visibility)
plt.title('Average Yearly Visibility')
plt.xlabel('Year')
plt.ylabel('Average Visibility (km)')
plt.xticks(rotation=45) 
plt.grid(True)
plt.show()
```
